# Tutorial for AirBnB data


Here is small sample code to try predict the price for Airbnb accomdation in Manchester. We will only use two features and simple linear regression.
Here we import what libaries we need including seaborn to make some nice plots along with some tasks from Scikit-learn, the main workhorse for your machine learning needs. We also import the data in and look at some basic characteristics usings pandas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
%matplotlib inline

initialdata = pd.read_csv('listings.csv')
print(initialdata.shape)
```

```python
initialdata.head()
```

```python
plt.figure(figsize=(13, 40))
(initialdata.isnull().mean(axis=0)*100).plot.barh()
plt.xlim(xmax=100)
plt.title("Missing values rate",fontsize=18)
plt.xlabel("percentage",fontsize=14)
```

For this simple tutorial will take only two features from the data set to predict accommodation prices, namely the number of reviews and the minimum number of nights one can stay at the accommodation. We now create two small dataframes, one containing only the features and the other only the prices.

```python
listofcols = ['number_of_reviews','minimum_nights'] #The features we will use
editeddata = initialdata[listofcols]
pricedata = initialdata['price']
```

We now need to split our data into a training dataset to train our model on and a test dataset to test our model predictions.

```python
X_train, X_test, y_train, y_test = train_test_split(editeddata, pricedata, test_size=0.15,random_state=0)
```

We initialise the linear regression model in scikit learn and fit it to the training data.

```python
from sklearn.linear_model import LinearRegression # Import linear regression
lm = LinearRegression() #Initialise 
lm.fit(X_train,y_train) #fit the model to the data
```

We will now get a measure on how accurate our model is using the root mean squared error which will be in units of gbp.

```python
print('Number of coefficents is', len(lm.coef_))
print('Estimated intercept coefficeint is', lm.intercept_)
print('Root mean squared error on the training data:', np.sqrt(mean_squared_error(y_train, lm.predict(X_train))))
print('Root mean squared error on the test data:', np.sqrt(mean_squared_error(y_test, lm.predict(X_test))))
```

From this result, we can expect the predicted price value to be off by 115 gbp on average. Not great! We can also try plot the residuals.

```python
plt.rcParams['axes.facecolor'] = 'white'
plt.scatter(lm.predict(X_train),lm.predict(X_train)- y_train, c='b', s=40,alpha=0.5)
plt.scatter(lm.predict(X_test),lm.predict(X_test)- y_test, c='g', s=40,alpha=0.5)
plt.axhline(0,color='r')
plt.ylabel('residuals')
plt.ylim(-1000,100)
plt.xlim(0,100)
plt.show()
```

We now try the same model but with a different random state. 

```python
X_train, X_test, y_train, y_test = train_test_split(editeddata, pricedata, test_size=0.15,random_state=2)
from sklearn.linear_model import LinearRegression # Import linear regression
lm = LinearRegression() #Initialise 
lm.fit(X_train,y_train) #fit the model to the data
print('Number of coefficents is', len(lm.coef_))
print('Estimated intercept coefficeint is', lm.intercept_)
print('Root Mean squared error on the training data:', np.sqrt(mean_squared_error(y_train, lm.predict(X_train))))
print('Root Mean squared error on the test data:', np.sqrt(mean_squared_error(y_test, lm.predict(X_test))))
```

There is significant variability between the two different random states when we are performing our simple cross-vaildation. In order to reduce variability, we need to perform multiple rounds of cross-validation using different partitions. The validation results of these partitions are then averaged over the rounds. We can use a method call K-Fold Cross Validation to help reduce this variability.
https://en.wikipedia.org/wiki/Cross-validation_(statistics)

We try testing a range of different folds, shown below:


```python
from sklearn.model_selection import cross_val_score, KFold

num_folds = [3, 5, 7, 9, 10, 11, 13, 15, 17, 19, 21, 23]

for fold in num_folds:
    kf = KFold(fold, shuffle=True, random_state=1) # Fold the data using a defined random state.
    model = LinearRegression()
    mses = cross_val_score(model, editeddata, pricedata, scoring="neg_mean_squared_error", cv=kf) # using original editeddata and pricedata
    rmses = [np.sqrt(np.absolute(mse)) for mse in mses]
    avg_rmse = np.mean(rmses)
    std_rmse = np.std(rmses)
    print(str(fold), "folds: ", "avg RMSE: ", str(avg_rmse), "std RMSE: ", str(std_rmse))
```

So we are finding that the model has an average error of approximately 95 gbp.
